```python
def process_events(
    events_df,
    reference_df,
    event_idx,
    window,
    nut_id,
    haz="flood",
    endpeaks_col="enddate",
    stpeaks_col="stardate",
    start_date_col="Start date",
    end_date_col="End date",
    id_col="eid"
):
    # Convert relevant columns to datetime (safely)
    events_df[endpeaks_col] = pd.to_datetime(events_df[endpeaks_col], errors='coerce')
    events_df[stpeaks_col] = pd.to_datetime(events_df[stpeaks_col], errors='coerce')
    reference_df[start_date_col] = pd.to_datetime(reference_df[start_date_col], errors='coerce')
    reference_df[end_date_col] = pd.to_datetime(reference_df[end_date_col], errors='coerce')

    # Extract reference event
    ref_start = reference_df.iloc[event_idx][start_date_col]
    ref_end = reference_df.iloc[event_idx][end_date_col]
    ref_id = reference_df.iloc[event_idx,][id_col]

    # Calculate date differences (equivalent to R’s d1, d2)
    d1 = (events_df[endpeaks_col] - ref_start).dt.days + 1
    d2 = (events_df[stpeaks_col] - ref_end).dt.days + 1

    # Determine which events to keep, hazard-specific logic
    if haz == "flood":
        keep_mask = (d1 <= -4) & (d1 >= -window)
    elif haz == "flmatch":
        keep_mask = (d2 <= window) & (d1 >= -window)
    elif haz in ["drought", "cold", "wind"]:
        keep_mask = (d2 <= 0) & (d1 >= -window)
    elif haz == "heat":
        keep_mask = (d1 <= 0) & (d1 >= -window)
    else:
        print("Invalid hazard type.")
        return pd.DataFrame()  # Empty dataframe

    # Keep the matching rows
    events_to_keep = events_df.loc[keep_mask].copy()

    if not events_to_keep.empty:
        events_to_keep["dtime1"] = d1[keep_mask].values
        events_to_keep["dtime2"] = d1[keep_mask].values
        events_to_keep["NUTID"] = nut_id
        events_to_keep["eventStart"] = ref_start
        events_to_keep["eventEnd"] = ref_end
        events_to_keep["eventID"] = ref_id
        return events_to_keep

    # If no matches found, return empty DataFrame
    return pd.DataFrame()


def elbow_finder(x_values, y_values, n_points=16):
    """
    Finds the 'elbow point' of a curve by computing the farthest distance 
    between the curve and a straight line connecting the min and max values.

    Parameters
    ----------
    x_values : array-like
        X coordinates of the curve.
    y_values : array-like
        Y coordinates of the curve.
    n_points : int
        Number of points to interpolate between min and max (default: 16).

    Returns
    -------
    tuple
        (x_max_dist, y_max_dist, xcmax, ycmax)
        where:
        - (x_max_dist, y_max_dist) = coordinates of the elbow on the curve
        - (xcmax, ycmax) = coordinates on the baseline line directly below the elbow
    """

    x_values = np.array(x_values, dtype=float)
    y_values = np.array(y_values, dtype=float)

    # Identify min-x and max-y points
    max_x_x = np.min(x_values)
    max_x_y = y_values[np.argmin(x_values)]
    max_y_y = np.max(y_values)
    max_y_x = x_values[np.argmax(y_values)]

    # Line between the two extreme points (max_x and max_y)
    line_x = np.linspace(max_y_x, max_x_x, n_points)
    line_y = np.linspace(max_y_y, max_x_y, n_points)

    # Normalize both line and curve
    xcurve = line_y / np.max(line_y)
    ycurve = line_x / np.max(line_x)
    x_values2 = x_values / np.max(x_values)
    y_values2 = y_values / np.max(y_values)

    # Compute minimal distances from curve to line
    distances = []
    x_min_dist = []
    y_min_dist = []

    for i in range(len(x_values2)):
        d = np.sqrt((x_values2[i] - xcurve)**2 + (y_values2[i] - ycurve)**2)
        min_idx = np.argmin(d)
        distances.append(np.min(d))
        x_min_dist.append(xcurve[min_idx])
        y_min_dist.append(ycurve[min_idx])

    distances = np.array(distances)
    x_min_dist = np.array(x_min_dist)
    y_min_dist = np.array(y_min_dist)

    # Find point with maximum distance
    max_idx = np.argmax(distances)
    x_max_dist = x_values[max_idx]
    y_max_dist = y_values[max_idx]
    xcmax = x_min_dist[max_idx]
    ycmax = y_min_dist[max_idx]

    return (x_max_dist, y_max_dist, xcmax, ycmax)
```

```python

fl_events = pd.read_csv(f"{hydroDir}/hazard_data/fl_nuts3_v1.csv", index_col=0)
fl_events['enddate'] = pd.to_datetime(fl_events['enddate'])
fl_events['stardate'] = fl_events['enddate'] - pd.to_timedelta(fl_events['duration'], unit='D')

dr_events = pd.read_csv(f"{hydroDir}/hazard_data/dr_nuts3_v1.csv", index_col=0)
dr_events['stardate'] = pd.to_datetime(dr_events['stardate'], format="%d/%m/%Y")
dr_events['enddate'] = pd.to_datetime(dr_events['enddate'], format="%d/%m/%Y")
dr_events['stardate'] = pd.to_datetime(dr_events['stardate'], format="%d/%m/%Y")
dr_events['enddate'] = pd.to_datetime(dr_events['enddate'], format="%d/%m/%Y")
dr_events['duration'] = (dr_events['enddate'] - dr_events['stardate']).dt.days+1
dr_events = dr_events[dr_events["duration"] >= 10]

hw_events = pd.read_csv(f"{hydroDir}/hazard_data/hw_nuts3_v1.csv")
hw_events['stardate'] = pd.to_datetime(hw_events['stardate'], format="%d/%m/%Y %H:%M", errors='coerce')
hw_events['enddate'] = pd.to_datetime(hw_events['enddate'], format="%d/%m/%Y %H:%M", errors='coerce')

cw_events = pd.read_csv(f"{hydroDir}/hazard_data/cw_nuts3_v1.csv")
cw_events['stardate'] = pd.to_datetime(cw_events['stardate'], format="%d/%m/%Y %H:%M", errors='coerce')
cw_events['enddate'] = pd.to_datetime(cw_events['enddate'], format="%d/%m/%Y %H:%M", errors='coerce')

ws_events = pd.read_csv(f"{hydroDir}/hazard_data/wgust_nuts3_995.csv")
ws_events['stardate'] = pd.to_datetime(ws_events['stardate'], format="%d/%m/%Y %H:%M", errors='coerce')
ws_events['enddate'] = pd.to_datetime(ws_events['enddate'], format="%d/%m/%Y %H:%M", errors='coerce')

```

```python


initF = True

if initF:
    HitF = []
    list_POTFl_large = []
    llpot = []
    llev = []

    for w in range(16):  # 0 to 15 inclusive
        wflood = w
        df_POTFlood_large = pd.DataFrame()
        Hanze_flood_M1 = pd.DataFrame()

        for NUTl in NutVector:
            High_events = fl_events[fl_events['NUT'] == NUTl]

            if not High_events.empty:
                # Match Hanze rows corresponding to this NUT
                matches = [NUTl in vec for vec in list_of_word_vectors]
                tm = [i for i, val in enumerate(matches) if val]
                Hanze_Nut = Hanze_flood.iloc[tm]
                Hanze_Nut = Hanze_Nut[Hanze_Nut['Year'] > 1980].copy()

                if not Hanze_Nut.empty:
                    Hanze_Nut['NUTl'] = NUTl
                    Hanze_flood_M1 = pd.concat([Hanze_flood_M1, Hanze_Nut], ignore_index=True)

                    # Compare each reference event
                    df_POTFlood_keep = pd.DataFrame()
                    Hanze_Nut['eid'] = Hanze_Nut['ID']

                    for e, eid in enumerate(Hanze_Nut['ID']):
                        POTh_keep = process_events(
                            events_df=High_events,
                            reference_df=Hanze_Nut,
                            event_idx=e,
                            window=wflood,
                            nut_id=NUTl,
                            haz="flmatch"
                        )
                        df_POTFlood_keep = pd.concat([df_POTFlood_keep, POTh_keep], ignore_index=True)

                    df_POTFlood_large = pd.concat([df_POTFlood_large, df_POTFlood_keep], ignore_index=True)

        if not Hanze_flood_M1.empty:
            Hanze_flood_M1['NEID'] = Hanze_flood_M1['ID'].astype(str) + Hanze_flood_M1['NUTl']

        # Aggregate matches per event-region
        Hanze_DfNUT = (
            df_POTFlood_large
            .groupby(['eventID', 'NUTID'])
            .size()
            .reset_index(name='time')
        )
        Hanze_DfNUT['NEID'] = Hanze_DfNUT['eventID'].astype(str) + Hanze_DfNUT['NUTID']

        Hanze_DFmNut = Hanze_DfNUT.merge(Hanze_flood_M1, on='NEID', how='inner')

        lev = len(Hanze_DFmNut)
        llev.append(lev)
        lpot = len(df_POTFlood_large)
        llpot.append(lpot)
        list_POTFl_large.append(df_POTFlood_large)

        # --- Summary stats
        Hanze_Dflood = df_POTFlood_large.groupby('eventID').size().reset_index(name='time')
        Hanze_DFmatch = Hanze_Dflood.merge(Hanze_flood_1980, left_on='eventID', right_on='ID', how='inner')

        total_hit = len(Hanze_DFmatch) / len(Hanze_flood_1980) * 100
        hit_river = (
            len(Hanze_DFmatch[Hanze_DFmatch['Type'] == 'River'])
            / len(Hanze_flood_1980[Hanze_flood_1980['Type'] == 'River'])
            * 100
        )
        hit_flash = (
            len(Hanze_DFmatch[Hanze_DFmatch['Type'] == 'Flash'])
            / len(Hanze_flood_1980[Hanze_flood_1980['Type'] == 'Flash'])
            * 100
        )

        HitF.append([w, total_hit, hit_river, hit_flash])

    HitF = pd.DataFrame(HitF, columns=['lag', 'Hit_total', 'Hit_river', 'Hit_flash'])

    # Ratio of new detected events
    nvar = np.diff(llev) / np.diff(llpot)
    nvar = np.append(1 - nvar, 0)

    # Find elbow
    optim = elbow_finder(HitF['lag'].values, nvar)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(HitF['lag'], HitF['Hit_total'], 'o-', label='All floods')
    plt.plot(HitF['lag'], HitF['Hit_river'], 'o-', label='Riverine')
    plt.plot(HitF['lag'], HitF['Hit_flash'], 'o-', label='Flash')
    plt.axvline(optim[0], color='gray', linestyle='--')
    plt.xlabel("Lag before and after event (days)")
    plt.ylabel("Hit rate (%)")
    plt.title("Optimal lag time for flood event matching")
    plt.ylim(25, 100)
    plt.legend()
    plt.show()
```

```python
w = 4
wflood = [w]

df_POTFlood_large = pd.DataFrame()
Hanze_flood_BN = pd.DataFrame()
list_POTFl_large = []

# Loop over NUTS regions
for NUTl in NutVector:
    High_events = fl_events[fl_events["NUT"] == NUTl].copy()

    if len(High_events) > 0:
        # Match NUTl with list_of_word_vectors
        matches = [NUTl in vector for vector in list_of_word_vectors]
        tm = [i for i, x in enumerate(matches) if x is True]

        Hanze_Nut = Hanze_flood.iloc[tm].copy()
        Hanze_Nut = Hanze_Nut[Hanze_Nut["Year"] > 1980].copy()

        if len(Hanze_Nut["Country code"]) > 0:
            Hanze_bNut = Hanze_Nut.copy()
            Hanze_bNut["NUTl"] = NUTl
            Hanze_flood_BN = pd.concat([Hanze_flood_BN, Hanze_bNut], ignore_index=True)

        # Check overlap with detected anomalies
        df_POTFlood_keep = pd.DataFrame()
        Hanze_Nut["eid"] = Hanze_Nut["ID"]

        for e in range(len(Hanze_Nut["ID"])):
            POTh_keep = process_events(
                events_df=High_events,
                reference_df=Hanze_Nut,
                event_idx=e,
                window=wflood[0],
                nut_id=NUTl,
                haz="flmatch"
            )
            df_POTFlood_keep = pd.concat([df_POTFlood_keep, POTh_keep], ignore_index=True)

        df_POTFlood_large = pd.concat([df_POTFlood_large, df_POTFlood_keep], ignore_index=True)

# Create NEID
Hanze_flood_BN["NEID"] = Hanze_flood_BN["ID"].astype(str) + Hanze_flood_BN["NUTl"].astype(str)

# Save to list
list_POTFl_large.append(df_POTFlood_large)

# ---- Summary of matched flood events ----
Hanze_Dflood = (
    df_POTFlood_large.groupby("eventID")
    .agg(time_count=("threshold", "count"))
    .reset_index()
    .rename(columns={"eventID": "NID"})
)

Hanze_DFmatch = Hanze_Dflood.merge(
    Hanze_flood_1980, left_on="NID", right_on="ID", how="inner"
)

Hit_tot = len(Hanze_DFmatch["NID"]) / len(Hanze_flood_1980["Country code"]) * 100
Hit_riv = (
    len(Hanze_DFmatch.loc[Hanze_DFmatch["Type"] == "River", "NID"])
    / len(Hanze_flood_1980.loc[Hanze_flood_1980["Type"] == "River", "Country code"])
    * 100
)
Hit_flash = (
    len(Hanze_DFmatch.loc[Hanze_DFmatch["Type"] == "Flash", "NID"])
    / len(Hanze_flood_1980.loc[Hanze_flood_1980["Type"] == "Flash", "Country code"])
    * 100
)

df_POTFlood = df_POTFlood_large.copy()

# ---- Aggregate by region for spatial success rate ----
Hanze_SDflood = (
    df_POTFlood.groupby(["NUT", "eventID"])
    .agg(he_count=("threshold", "count"))
    .reset_index()
)

Hanze_SDflood = (
    Hanze_SDflood.groupby("NUT")
    .agg(he_count=("he_count", "count"))
    .reset_index()
)

Hanze_SOflood = (
    Hanze_flood_BN.groupby("NUTl")
    .agg(he_count=("Year", "count"))
    .reset_index()
    .rename(columns={"NUTl": "NUT"})
)

Hanze_spRat = pd.merge(Hanze_SOflood, Hanze_SDflood, on="NUT", how="outer")
Hanze_spRat["succerat"] = Hanze_spRat["he_count_y"] / Hanze_spRat["he_count_x"] * 100

```

```python

ICW, IHW, IWS, IFlood, IDrought = 14, 14, 56, 56, 56
short, medium, long = 14, 28, 56
windows = ["informed", "short", "medium", "long"]

list_POTn_large = []
list_POTx_large = []
list_POTh_large = []
list_POTl_large = []
list_POTw_large = []

for w in windows:
    # Assign windows for each hazard type
    if w == "informed":
        wflood, wdrought, wheat, wcold, wwind = IFlood, IDrought, IHW, ICW, IWS
    elif w == "long":
        wflood = wdrought = wheat = wcold = wwind = long
    elif w == "medium":
        wflood = wdrought = wheat = wcold = wwind = medium
    elif w == "short":
        wflood = wdrought = wheat = wcold = wwind = short

    # Initialize DataFrames to store detected overlaps
    df_POTn_large_w = pd.DataFrame()
    df_POTx_large_w = pd.DataFrame()
    df_POTh_large_w = pd.DataFrame()
    df_POTl_large_w = pd.DataFrame()
    df_POTw_large_w = pd.DataFrame()

    for NUTl in NutVector:
        # Subset events per region
        High_events = fl_events[fl_events["NUT"] == NUTl]
        if not High_events.empty:
            Low_events = dr_events[dr_events["NUT"] == NUTl]
            HW_events = hw_events[hw_events["nuts_id"] == NUTl]
            CW_events = cw_events[cw_events["nuts_id"] == NUTl]
            WS_events = ws_events[ws_events["nuts_id"] == NUTl]

            # Match Hanze rows for this NUT
            matches = [NUTl in vec for vec in list_of_word_vectors]
            tm = [i for i, val in enumerate(matches) if val]
            Hanze_Nut = Hanze_flood.iloc[tm]
            Hanze_Nut = Hanze_Nut[Hanze_Nut["Year"] > 1980].copy()
            Hanze_Nut["eid"] = Hanze_Nut["ID"]

            # Initialize per-hazard DataFrames
            df_POTh_keep = pd.DataFrame()
            df_POTl_keep = pd.DataFrame()
            df_POTx_keep = pd.DataFrame()
            df_POTn_keep = pd.DataFrame()
            df_POTw_keep = pd.DataFrame()

            for e, eid in enumerate(Hanze_Nut["ID"]):
                # --- Flood sequence ---
                POTh_keep = process_events(
                    events_df=High_events,
                    reference_df=Hanze_Nut,
                    event_idx=e,
                    window=wflood,
                    nut_id=NUTl,
                    haz="flood",
                )
                df_POTh_keep = pd.concat([df_POTh_keep, POTh_keep], ignore_index=True)

                # --- Drought-Flood sequence ---
                POTl_keep = process_events(
                    events_df=Low_events,
                    reference_df=Hanze_Nut,
                    event_idx=e,
                    window=wdrought,
                    nut_id=NUTl,
                    haz="drought",
                )
                df_POTl_keep = pd.concat([df_POTl_keep, POTl_keep], ignore_index=True)

                # --- Hot-Wet sequence ---
                POTx_keep = process_events(
                    events_df=HW_events,
                    reference_df=Hanze_Nut,
                    event_idx=e,
                    window=wheat,
                    nut_id=NUTl,
                    haz="heat",
                )
                df_POTx_keep = pd.concat([df_POTx_keep, POTx_keep], ignore_index=True)

                # --- Cold-Flood sequence ---
                POTn_keep = process_events(
                    events_df=CW_events,
                    reference_df=Hanze_Nut,
                    event_idx=e,
                    window=wcold,
                    nut_id=NUTl,
                    haz="cold",
                )
                df_POTn_keep = pd.concat([df_POTn_keep, POTn_keep], ignore_index=True)

                # --- Wind-Flood sequence ---
                POTw_keep = process_events(
                    events_df=WS_events,
                    reference_df=Hanze_Nut,
                    event_idx=e,
                    window=wwind,
                    nut_id=NUTl,
                    haz="wind",
                )
                df_POTw_keep = pd.concat([df_POTw_keep, POTw_keep], ignore_index=True)

            # Append region-level results
            df_POTh_large_w = pd.concat([df_POTh_large_w, df_POTh_keep], ignore_index=True)
            df_POTl_large_w = pd.concat([df_POTl_large_w, df_POTl_keep], ignore_index=True)
            df_POTx_large_w = pd.concat([df_POTx_large_w, df_POTx_keep], ignore_index=True)
            df_POTn_large_w = pd.concat([df_POTn_large_w, df_POTn_keep], ignore_index=True)
            df_POTw_large_w = pd.concat([df_POTw_large_w, df_POTw_keep], ignore_index=True)

    # Add NEID column per hazard
    for df in [df_POTh_large_w, df_POTl_large_w, df_POTx_large_w, df_POTn_large_w, df_POTw_large_w]:
        df["NEID"] = df["eventID"].astype(str) + df["NUTID"]

    # Append to master lists
    list_POTh_large.append(df_POTh_large_w)
    list_POTl_large.append(df_POTl_large_w)
    list_POTx_large.append(df_POTx_large_w)
    list_POTn_large.append(df_POTn_large_w)
    list_POTw_large.append(df_POTw_large_w)
```

```python

output_dir = "../plots/"
os.makedirs(output_dir, exist_ok=True)

plot1 = []
list_eventc2 = []
list_call = []

for w, current_window in enumerate(windows, start=0):
    # Merge modeled events with observed Hanze floods
    Events_prefflood = pd.merge(list_POTh_large[w], Hanze_flood_1980, left_on="eventID", right_on="ID")
    Events_precoflood = pd.merge(list_POTl_large[w], Hanze_flood_1980, left_on="eventID", right_on="ID")
    Events_prehflood = pd.merge(list_POTx_large[w], Hanze_flood_1980, left_on="eventID", right_on="ID")
    Events_precflood = pd.merge(list_POTn_large[w], Hanze_flood_1980, left_on="eventID", right_on="ID")
    Events_prewflood = pd.merge(list_POTw_large[w], Hanze_flood_1980, left_on="eventID", right_on="ID")

    # Prepare 'time' column for heat, cold, wind
    for df in [Events_prehflood, Events_precflood, Events_prewflood]:
        df['time'] = pd.to_datetime(df['stardate'])

    # ---- Flood events ----
    C_flood_ev = Events_prefflood.copy()
    C_flood_ev['stdate'] = pd.to_datetime(C_flood_ev['stpeaks'])
    kept_col = ["NUT", "dtime1", "duration", "peaks", "time", "eventID", "NEID"]
    C_flood_ev = C_flood_ev[kept_col]
    C_flood_ev['peaks'] = (C_flood_ev['peaks'] - 0.98) / 0.02
    C_flood_ev['time'] = pd.to_datetime(C_flood_ev['time'])
    
    # ---- Cold events ----
    C_cold_ev = Events_precflood.copy()
    C_cold_ev['duration'] = C_cold_ev['intensity']
    C_cold_ev['NUT'] = C_cold_ev['nuts_id']
    C_cold_ev['peaks'] = 1
    C_cold_ev.loc[C_cold_ev['dtime1'] > 0, 'dtime1'] = 0
    C_cold_ev['time'] = pd.to_datetime(C_cold_ev['time'])
    # remove cold events during warm season (May-Oct)
    C_cold_ev = C_cold_ev[~C_cold_ev['time'].dt.month.between(5,10)]

    # ---- Heat events ----
    C_heat_ev = Events_prehflood.copy()
    C_heat_ev['duration'] = C_heat_ev['intensity']
    C_heat_ev['NUT'] = C_heat_ev['nuts_id']
    C_heat_ev['peaks'] = 1
    C_heat_ev['time'] = pd.to_datetime(C_heat_ev['time'])
    # remove heat events during cold season (Nov-Apr)
    C_heat_ev = C_heat_ev[C_heat_ev['time'].dt.month.between(5,10)]

    # ---- Wind events ----
    C_wind_ev = Events_prewflood.copy()
    C_wind_ev['duration'] = (pd.to_datetime(C_wind_ev['enddate']) - pd.to_datetime(C_wind_ev['stardate'])).dt.days
    C_wind_ev['NUT'] = C_wind_ev['nuts_id']
    C_wind_ev['peaks'] = C_wind_ev['intensity']
    C_wind_ev.loc[C_wind_ev['dtime1'] > 0, 'dtime1'] = 0
    C_wind_ev['time'] = pd.to_datetime(C_wind_ev['time'])

    # ---- Drought events ----
    C_drought_ev = Events_precoflood.copy()
    C_drought_ev['peaks'] = ((1 - C_drought_ev['low_peaks']) - 0.95) / 0.05
    C_drought_ev['duration'] = C_drought_ev['cluster_length']
    C_drought_ev.loc[C_drought_ev['dtime1'] > 0, 'dtime1'] = 0
    C_drought_ev['time'] = pd.to_datetime(C_drought_ev['time'])

    # ---- Add event types ----
    for df, evtype in zip([C_flood_ev, C_drought_ev, C_heat_ev, C_cold_ev, C_wind_ev],
                          ["flood", "drought", "heat", "cold", "wind"]):
        df['evtype'] = evtype
        # Align columns
        df = df[C_flood_ev.columns]

    # ---- Combine all hazards ----
    C_all_haz = pd.concat([C_flood_ev, C_drought_ev, C_heat_ev, C_cold_ev, C_wind_ev], ignore_index=True)
    list_call.append(C_all_haz)

    # ---- Classify events by number of hazards ----
    all_events = Hanze_flood_1980['ID'].unique()
    mh_events = C_all_haz['eventID'].unique()
    single_events = np.setdiff1d(all_events, mh_events)

    evtypes = ["drought","flood","heat","cold","wind","single"]
    events_class_list = []

    for evsel in all_events:
        C_all_s = C_all_haz[C_all_haz['eventID'] == evsel]
        cls = {'eventID': evsel, 'predrought':0,'preflood':0,'preheat':0,'precold':0,'prewind':0,'single':0}
        if C_all_s.shape[0] == 0:
            cls['single'] = 1
        else:
            for ev in C_all_s['evtype'].unique():
                key = f'pre{ev}' if ev != "single" else "single"
                cls[key] = 1
        cls['nh'] = cls['predrought'] + cls['preflood'] + cls['preheat'] + cls['precold'] + cls['prewind']
        events_class_list.append(cls)

    events_class2 = pd.DataFrame(events_class_list)

    # ---- Assign names to multi-hazard categories ----
    def assign_event_name(row):
        nh = row['nh']
        if nh == 0: return "single-flood"
        if nh == 1:
            if row['predrought']: return "drought-flood"
            if row['preflood']: return "wet-sequence"
            if row['preheat']: return "heat-flood"
            if row['precold']: return "cold-flood"
            if row['prewind']: return "wind-flood"
        if nh == 2:
            if row['predrought'] and row['preflood']: return "drought-wet-sequence"
            if row['predrought'] and row['preheat']: return "heat-drought-flood"
            if row['preflood'] and row['preheat']: return "heat-wet-sequence"
            if row['preflood'] and row['precold']: return "cold-wet-sequence"
            if row['preheat'] and row['precold']: return "cold-heat-wet"
            if row['predrought'] and row['precold']: return "cold-drought-flood"
            if row['predrought'] and row['prewind']: return "wind-drought-flood"
            if row['preflood'] and row['prewind']: return "wind-wet-sequence"
            if row['precold'] and row['prewind']: return "wind-cold-flood"
            if row['preheat'] and row['prewind']: return "wind-heat-flood"
        if nh == 3:
            if row['predrought'] and row['preflood'] and row['preheat']: return "heat-drought-to-wet-sequence"
            if row['predrought'] and row['preflood'] and row['precold']: return "cold-drought-to-wet-sequence"
            if row['preheat'] and row['preflood'] and row['precold']: return "cold-heat-wet-sequence"
            if row['prewind'] and row['preheat'] and row['precold']: return "wind-cold-heat-flood"
            if row['predrought'] and row['prewind'] and row['precold']: return "wind-cold-drought-flood"
            if row['predrought'] and row['preheat'] and row['prewind']: return "wind-heat-drought-flood"
            if row['preflood'] and row['preheat'] and row['prewind']: return "wind-heat-wet-sequence"
            if row['preflood'] and row['precold'] and row['prewind']: return "wind-cold-wet-sequence"
            if row['preflood'] and row['predrought'] and row['prewind']: return "wind-drought-to-wet-sequence"
            if row['precold'] and row['preheat'] and row['predrought']: return "cold-heat-drought-flood"
        if nh == 4:
            if row['precold'] and row['predrought'] and row['preflood'] and row['preheat']:
                return "heat-cold-drought-to-wet-sequence"
            if row['precold'] and row['predrought'] and row['preflood'] and row['prewind']:
                return "wind-cold-drought-to-wet-sequence"
            if row['prewind'] and row['predrought'] and row['preflood'] and row['preheat']:
                return "wind-heat-drought-to-wet-sequence"
            if row['prewind'] and row['precold'] and row['preflood'] and row['preheat']:
                return "wind-heat-cold-wet-sequence"
            if row['prewind'] and row['predrought'] and row['precold'] and row['preheat']:
                return "wind-heat-cold-drought-flood"
        if nh == 5: return "wind-heat-cold-drought-to-wet-sequence"
        return None

    events_class2['event_name'] = events_class2.apply(assign_event_name, axis=1)
    events_class2['nhcat'] = events_class2['nh'].astype(str)
    list_eventc2.append(events_class2)

    # ---- Order the categories based on frequency ----
    ordered_event_names = events_class2['event_name'].value_counts().index

    # ---- Define a custom color palette ----
    custom_palette = {
        '0': '#f0f0f0',  # light grey for single hazards
        '1': '#ffecb3',  # light yellow for one additional hazard
        '2': '#ffb347',  # orange for two additional hazards
        '3': '#ff7f50',  # coral for three additional hazards
        '4': '#ff4500',  # red for four additional hazards
        '5': '#b22222'   # dark red for five additional hazards
    }

    # ---- Plot barchart ----
    plt.figure(figsize=(10, 6), dpi=300)  # Setting high DPI for publication quality
    sns.countplot(data=events_class2, y='event_name', hue='nhcat', order=ordered_event_names, palette=custom_palette)
    plt.xlabel("Number of events")
    plt.ylabel("Compound event name")
    plt.title(f"Compound hazard events for {current_window} window")
    plt.legend(title="Number of compounding hazards", bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=3)
    plt.tight_layout()
    
    # Save the plot
    plot_filename = f"{output_dir}compound_hazard_events_{current_window}.png"
    plt.savefig(plot_filename, bbox_inches='tight')
    plt.close()
    
    plot1.append(plot_filename)

print("Plots saved in:", output_dir)

```

```python
Hanze_events_RP = pd.read_csv(f"{hydroDir}/HANZE/Attribution_losses_under_factual_scenario_river_with_RP.csv")
Hanze_events_RP['NEID'] = Hanze_events_RP['HANZE_ID'].astype(str) + Hanze_events_RP['NUTS3'].astype(str)
unikid = Hanze_events_RP['HANZE_ID'].unique()
Hanze_events_RP2 = pd.DataFrame()

for evx in unikid:
    # Subset dataframe for current HANZE_ID
    sev = Hanze_events_RP[Hanze_events_RP['HANZE_ID'] == evx]

    # Calculate ratios for Economic_loss
    sumloss = sev['Economic_loss'].sum()
    sev['ratioloss'] = (sev['Economic_loss'] / sumloss).round(2)

    # Calculate ratios for Fatalities
    sumfat = sev['Fatalities'].sum()
    sev['ratiofat'] = (sev['Fatalities'] / sumfat).round(2)

    # Calculate ratios for Persons_affected
    sumaf = sev['Persons_affected'].sum()
    sev['ratioaf'] = (sev['Persons_affected'] / sumaf).round(2)

    # Append the processed DataFrame to the results
    Hanze_events_RP2 = pd.concat([Hanze_events_RP2, sev], ignore_index=True)

# Drop unwanted columns (assuming columns to drop are 1, 10, 11, and 12 based on zero-indexing in Python)
Hanze_events_RP2.drop(Hanze_events_RP2.columns[[0, 9, 10, 11]], axis=1, inplace=True)

# Assuming Hanze_flood_BN is already loaded, map ReturnPeriod using NEID
rpm = Hanze_flood_BN['NEID'].map(Hanze_events_RP.set_index('NEID')['ReturnPeriod'])
Hanze_flood_BN['RPflood'] = rpm
ids_BN = set(Hanze_flood_BN['ID'].unique())
ids_1980 = set(Hanze_flood_1980['ID'].unique())
unique_ids_BN = ids_BN - ids_1980
rows_with_unique_ids = Hanze_flood_BN[Hanze_flood_BN['ID'].isin(unique_ids_BN)]
```

```python
RegioEv_data = list_call[0]

Hanze_flood_BN = Hanze_flood_BN[Hanze_flood_BN.Type != "Coastal"]
all_Revents = Hanze_flood_BN['NEID'].unique()

# Hazard categories (must match evtype values)
evtypes = ["drought", "flood", "heat", "cold", "wind"]

# Pre-size list: event ID + NUTSID + 5 hazards * 3 metrics = 17 columns
events_class = []

for evi in all_Revents:

    # Subset RegioEv_data for this event
    myev = RegioEv_data[RegioEv_data['NEID'] == evi]

    # Prepare row buffer
    cls = [None] * 17
    cls[0] = evi  # eventxNUT_ID

    # ✅ Always fill NUTSID from Hanze_flood_BN
    rowH = Hanze_flood_BN.loc[Hanze_flood_BN['NEID'] == evi]
    if not rowH.empty:
        cls[1] = rowH['NUTl'].iloc[0]
    else:
        cls[1] = None   # Should not happen unless Hanze is inconsistent

    # Hazards present for this event
    uh = myev['evtype'].unique()

    for hz in uh:

        # Skip unexpected hazards
        if hz not in evtypes:
            continue

        hz_idx = evtypes.index(hz)

        # Extract this hazard subset
        myehz = myev[myev['evtype'] == hz].copy()
        if myehz.empty:
            continue

        # Sort by time for grouping
        myehz = myehz.sort_values('dtime1')
        myehz['grp'] = 1
        grp = 1

        # ✅ Correct time-lag grouping logic
        if len(myehz) > 1:
            for dt in range(1, len(myehz)):
                diff = myehz.iloc[dt]['dtime1'] - myehz.iloc[dt-1]['dtime1']
                if diff >= 2:
                    grp += 1
                myehz.iat[dt, myehz.columns.get_loc('grp')] = grp

        # Number of distinct peaks (groups)
        lev = myehz['grp'].nunique()

        # Intensity = sum(peaks * duration)
        intensity = (myehz['peaks'] * myehz['duration']).sum()

        # ✅ Lag (closest to zero)
        if len(myehz) > 1:
            md_idx = myehz['dtime1'].abs().idxmin()
            lag_val = myehz.loc[md_idx, 'dtime1']
        else:
            lag_val = myehz.iloc[0]['dtime1']

        # Fill row
        cls[2 + hz_idx * 3] = lev
        cls[3 + hz_idx * 3] = intensity
        cls[4 + hz_idx * 3] = lag_val

    events_class.append(cls)

# Build dataframe
events_class_df = pd.DataFrame(events_class, columns=[
    "eventxNUT_ID", "NUTSID",
    "drought.n", "drought.i", "drought.lag",
    "flood.n", "flood.i", "flood.lag",
    "heat.n", "heat.i", "heat.lag",
    "cold.n", "cold.i", "cold.lag",
    "wind.n", "wind.i", "wind.lag"
])

# Convert hazard metrics to numeric
events_class_df.iloc[:, 2:17] = events_class_df.iloc[:, 2:17].apply(pd.to_numeric, errors='coerce')

# Remove coastal flooding events
Hanze_flood_BN2 = Hanze_flood_BN[Hanze_flood_BN['Type'] != "Coastal"]

# Final merge
RegioEvents = events_class_df.merge(
    Hanze_flood_BN2,
    left_on="eventxNUT_ID",
    right_on="NEID",
    how="inner"
)

```

```python
# Define function to create 10-year intervals
def create_10_year_intervals(start, end):
    intervals = []
    current_start = start
    while current_start + 9 <= end:
        intervals.append((current_start, current_start + 9))
        current_start += 10
    return intervals

# Create 10-year intervals from 1980 to 2020 (as an example)
ten_year_periods = create_10_year_intervals(1980, 2020)

# Calculate averages for each 10-year period
averages_list = []
for start_year, end_year in ten_year_periods:
    averages = calculate_averages(pivot_counts.reset_index(), start_year, end_year)
    averages_list.append((start_year, end_year, averages))

# Optionally, print the results
for start_year, end_year, averages in averages_list:
    print(f"Averages from {start_year} to {end_year}:\n{averages}\n")
```

```python

n_perm = 100000
loss_col = 'Losses (2020 euro)'
hazards = ["predrought", "preflood", "preheat", "precold", "prewind"]
random_seed = 42
output_png = '../plots/permutation_violin_plot_100k.png'

# --- Initialize RNG ---
rng = np.random.default_rng(random_seed)

# --- Prepare a separate DataFrame for analysis ---
Events_class_clean = Events_class.dropna(subset=[loss_col]).copy()
Events_class_clean['log_loss'] = np.log10(Events_class_clean[loss_col].clip(lower=1))

# --- Define "single" baseline ---
Events_class_clean['is_single'] = Events_class_clean['single'] == 1

# --- Benjamini–Hochberg FDR ---
def bh_fdr(pvals):
    p = np.asarray(pvals)
    n = len(p)
    order = np.argsort(p)
    ranked = p[order]
    q = np.empty(n)
    min_adj = 1.0
    for i in range(n-1, -1, -1):
        adj = ranked[i] * n / (i + 1)
        min_adj = min(min_adj, adj)
        q[order[i]] = min_adj
    q[q>1] = 1.0
    return q

def perm_test_hazard(df, hazard_col, n_perm=100000, rng=None):
    # Subset to relevant events only
    df_union = df[(df[hazard_col] == 1) | (df['single'] == 1)].copy().reset_index(drop=True)
    
    nA = (df_union[hazard_col] == 1).sum()
    nB = (df_union['single'] == 1).sum()
    
    # Debugging information
    print(f"Testing {hazard_col}: {nA} hazard events, {nB} single events")

    if nA == 0 or nB == 0:
        return dict(hazard=hazard_col, nA=nA, nB=nB,
                    obs_diff=np.nan, p_value=np.nan,
                    perm_diffs=np.array([]), ci=(np.nan, np.nan))
    
    # Observed mean difference
    meanA = df_union.loc[df_union[hazard_col] == 1, 'log_loss'].mean()
    meanB = df_union.loc[df_union['single'] == 1, 'log_loss'].mean()
    obs_diff = meanA - meanB

    # --- Permutation ---
    losses = df_union['log_loss'].values
    N = len(losses)
    perm_diffs = np.empty(n_perm)
    idx = np.arange(N)
    for i in range(n_perm):
        # Randomly choose nA indices for the hazard group
        idxA = rng.choice(idx, size=nA, replace=False)
        maskA = np.zeros(N, dtype=bool)
        maskA[idxA] = True
        perm_diffs[i] = losses[maskA].mean() - losses[~maskA].mean()
    
    # --- Empirical one-sided p-value ---
    p_value = np.mean(perm_diffs >= obs_diff)
    ci = (np.percentile(perm_diffs, 2.5), np.percentile(perm_diffs, 97.5))
    
    return dict(
        hazard=hazard_col, nA=nA, nB=nB,
        meanA=meanA, meanB=meanB, obs_diff=obs_diff,
        p_value=p_value, perm_diffs=perm_diffs, ci=ci
    )

# --- Run permutation tests ---
results = []
for h in hazards:
    print(f"Running {h} ({n_perm:,} permutations)...")
    res = perm_test_hazard(Events_class_clean, h, n_perm=n_perm, rng=rng)
    results.append(res)

# --- Summarize results ---
summary = pd.DataFrame([{
    'hazard': r['hazard'],
    'n_hazard': r['nA'],
    'n_single': r['nB'],
    'mean_log_hazard': r['meanA'],
    'mean_log_single': r['meanB'],
    'obs_mean_diff': r['obs_diff'],
    'perm_p_value': r['p_value'],
    'perm_ci_lower': r['ci'][0],
    'perm_ci_upper': r['ci'][1]
} for r in results])
summary['bh_q'] = bh_fdr(summary['perm_p_value'].fillna(1.0))
summary = summary.sort_values('perm_p_value').reset_index(drop=True)

# --- Publication-ready violin plot ---
pairs, labels = [], []
for h in hazards:
    maskA = Events_class_clean[h] == 1
    maskB = Events_class_clean['is_single'] == True
    if maskA.sum() == 0 or maskB.sum() == 0:
        continue
    pairs.append((Events_class_clean.loc[maskA, 'log_loss'].values,
                  Events_class_clean.loc[maskB, 'log_loss'].values))
    labels.append(h)

all_data, positions = [], []
pos = 1
for (a, b) in pairs:
    all_data.append(a); positions.append(pos); pos += 1
    all_data.append(b); positions.append(pos); pos += 1

fig, ax = plt.subplots(figsize=(max(8, len(positions)*0.6), 6))
vp = ax.violinplot(all_data, positions=positions, showmeans=False, showmedians=True)

# Add means and connecting lines
for i, (a, b) in enumerate(pairs):
    pos_h, pos_s = 2*i + 1, 2*i + 2
    mean_h, mean_s = a.mean(), b.mean()
    ax.plot([pos_h], [mean_h], marker='o', markersize=6)
    ax.plot([pos_s], [mean_s], marker='o', markersize=6)
    ax.plot([pos_h, pos_s], [mean_h, mean_s], linestyle='--', linewidth=0.8)

# Annotate p-values and q-values
ymax = max([np.max(arr) for arr in all_data])
yrange = ymax - min([np.min(arr) for arr in all_data])
annot_y = ymax + 0.08 * yrange
for i, h in enumerate(labels):
    row = summary.loc[summary['hazard'] == h].iloc[0]
    text = f"p={row['perm_p_value']:.3g}\nq={row['bh_q']:.3g}"
    ax.text(2*i+1.5, annot_y, text, ha='center', va='bottom', fontsize=9)

# Format axes
ax.set_xticks([2*i+1.5 for i in range(len(labels))])
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=10)
ax.set_ylabel('log10(Losses, 2020 euro)', fontsize=11)
ax.set_title('Hazard vs Single — Permutation Test (mean log-loss difference)', fontsize=12)
ax.set_xlim(0.5, pos-0.5)
plt.tight_layout()

# Save and show
os.makedirs(os.path.dirname(output_png), exist_ok=True)
plt.savefig(output_png, dpi=300)
plt.show()

# --- Display summary ---
print(f"\nSaved violin plot to: {output_png}\n")
print(summary.to_string(index=False))
```

```python

results_pot_log = []

# Set percentile threshold for POT
threshold_percentile = 75  # Can adjust (75th-90th)

for h in hazards:
    mask_h = Events_class_clean[h] == 1
    mask_s = Events_class_clean['is_single'] == True

    # Hazard-specific log_loss
    log_loss_h = Events_class_clean.loc[mask_h, 'log_loss'].values
    log_loss_s = Events_class_clean.loc[mask_s, 'log_loss'].values

    # Hazard-specific thresholds
    if len(log_loss_h) == 0 or len(log_loss_s) == 0:
        print(f"No data for hazard {h}. Skipping...")
        continue

    threshold_h = np.percentile(log_loss_h, threshold_percentile)
    threshold_s = np.percentile(log_loss_s, threshold_percentile)

    # Select exceedances
    extreme_h = log_loss_h[log_loss_h > threshold_h]
    extreme_s = log_loss_s[log_loss_s > threshold_s]

    if len(extreme_h) < 5 or len(extreme_s) < 5:
        print(f"Too few extreme points for {h}. Skipping...")
        continue

    # Fit GPD
    params_h = genpareto.fit(extreme_h, loc=threshold_h)
    params_s = genpareto.fit(extreme_s, loc=threshold_s)

    results_pot_log.append({
        'hazard': h,
        'n_exceed_hazard': len(extreme_h),
        'n_exceed_single': len(extreme_s),
        'params_hazard': params_h,
        'params_single': params_s
    })

# Convert to DataFrame
summary_pot_log = pd.DataFrame(results_pot_log)
print(summary_pot_log.to_string(index=False))

# =========================
# Visualization
# =========================

# Plot all hazards in subplots
fig, axes = plt.subplots(len(results_pot_log), 1, figsize=(10, 5*len(results_pot_log)))

if len(results_pot_log) == 1:
    axes = [axes]

for ax, row in zip(axes, results_pot_log):
    h = row['hazard']
    mask_h = Events_class_clean[h] == 1
    log_loss_h = Events_class_clean.loc[mask_h, 'log_loss'].values
    threshold_h = np.percentile(log_loss_h, threshold_percentile)
    extreme_h = log_loss_h[log_loss_h > threshold_h]

    # Get GPD parameters
    shape, loc, scale = row['params_hazard']
    x_max = loc - scale/shape if shape < 0 else extreme_h.max()
    x = np.linspace(threshold_h, x_max, 100)
    pdf_fitted = genpareto.pdf(x, shape, loc=loc, scale=scale)

    # Plot
    ax.hist(extreme_h, bins=20, density=True, alpha=0.5, label='Empirical')
    ax.plot(x, pdf_fitted, 'r-', lw=2, label='Fitted GPD')
    ax.set_xlabel('Log Loss')
    ax.set_ylabel('Density')
    ax.set_title(f'Extreme Value Analysis (Log Loss) - {h}')
    ax.legend()

plt.tight_layout()
plt.show()
```
